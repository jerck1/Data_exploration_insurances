### Triángulo de pérdidas

Creamos el triángulo de pérdidas para mirar como han evolucionado los pagos de las pólizas a partir del año en que ocurre el accidente, se toma como origen el año del accidente y como development el año en que se evalua el pago de la póliza. Se crean triángulos para cada una de las variables tanto de pérdidas (pólizas pagadas) como de ganancias (polizas ganadas en las compañías). El proceso se puede clasificar para cada compañía, por lo que se escoge como índice el nombre de las compañías GRNAME

```python
# #Ejemplo con una variable
# triangle_ex = cl.Triangle(
#     data[data.columns[:6]],
#     origin="AccidentYear",
#     development="DevelopmentYear",
#     columns='IncurLoss_D',
#     index=["GRNAME"],
#     cumulative=True,
# )
# triangle_ex
```

```python
# data[data.columns[:4]]
```

```python
triangle = cl.Triangle(
    data,
    origin="AccidentYear", # filas
    development="DevelopmentYear", # columnas
    columns=data.columns[5:], # tomamos en cuenta solamente las columnas de ganancias y pérdidas
    index=["GRNAME"], # nombres de las compañías
    cumulative=True, # triángulo de pérdidas acumulado o no
)
triangle #objeto de pandas
```

Obtenemos 132x8 triángulos (132 empresas, 8 variables/columnas) ,cada uno de tamaño 19x19.

Ejemplo de algunas de las compañías que se convierten en índices del triángulo de pérdidas

```python
triangle.index.head()
```

```python
print("Estructura del Chainladder triangle:", type(triangle.values))
 # suma de todas las pérdidas paara todas las empresas
print("Suma de los valores de la tabla de datos:", np.nansum(triangle.values))
```

### Creamos dos nuevas variables en el triángulo de pérdidas.

"CaseincurLoss": Esta nueva variable se crea restando el valor de "BulkLoss_D" de "IncurLoss_D". Básicamente, calcula la diferencia entre las pérdidas incurridas y las pérdidas globales  para cada punto de datos correspondiente en el conjunto de datos. Nos da una idea de cuales fueron aquéllas pérdidas adicionales a lo que se pagó en la prima

"PaidToInc": Esta nueva variable se crea dividiendo el valor de "CumPaidLoss_D" por "IncurLoss_D". Básicamente, calcula la proporción de pérdidas incurridas que se han pagado hasta un punto específico en el tiempo.

```python
triangle["CaseIncurLoss"] = triangle["IncurLoss_D"] - triangle["BulkLoss_D"] 
triangle["PaidToInc"] = triangle["CumPaidLoss_D"] / triangle["IncurLoss_D"] 
triangle[["CaseIncurLoss", "PaidToInc"]]
```

Ejemplo para la primer empresa

```python
triangle.loc[freq.index[0]].iloc[-1]['CaseIncurLoss']
```

Para la 1er empresa hay una optimización de los gastos de las primas incurridas con el paso de los años, reduciéndose en varios órdenes de magnitud
